```python
genres = df.Genre.isin(['Sports','Action','Shooter'])

# sales greater than 2 million
sales_2M = df['Global_Sales'] >= 2
```

```python
# Between
# specify a range -- same as between keyword in SQL

# games sales that are between 2 million and 10 million

sales_2M_10M = df['Global_Sales'].between(2,10)

# 90s games -- from 1990 to 1999 and so on

y_90s = df['Year'].between(1990,1999)
y_00s = df['Year'].between(2000,2009)
y_10s = df['Year'].between(2010,2016)

df.loc[sales_2M_10M & y_90s , "Genre"]

```

```python
# rename
df.rename({'EU_Sales': 'Europe_Sales',
           'NA_Sales': 'NorthAmerica_Sales',
           'JP_Sales': 'Japan_Sales'},
          axis=1,
           inplace=True)
```

## aggregation functions


```python
print("Min Year: ",df.Year.min())
print("Max Year: ",df.Year.max())
print("Median Year: ",df.Year.median())
print("Number of rows in a Year: ",df.Year.count())
print("Sum of Global Sales: ",df.Global_Sales.sum())

```

```python
# pd.cut 
# helps us bucketing of continuos data 
# also called as binning , bins

pd.cut(x = df["Year"],bins=5)
```

```python
# we can create a new column called Decade and assign it the bins we created

df['Decade'] = pd.cut(x = df['Year'], 
                      bins = [1980, 1989, 1999, 2009, 2016], 
                      labels = ["80s","90s","00s","10s"],
                      include_lowest=True)
```

```python
pd.cut(x = df["Year"],bins=5).value_counts()
```

```python
# or we can also bin with years where we specify the range

pd.cut(x = df['Year'], bins = [1980, 1989, 1999, 2009, 2016], labels = ["80s","90s","00s","10s"]).value_counts()
```

```python
# groupby

# summarize data across different groups, helps in breaking the df

# gives average sales by Genre in North America

df.groupby('Genre')['NorthAmerica_Sales'].mean()
 
# Genre is the index against which the average values are mapped 
```

```python
sales_by_Genre = df.groupby('Genre')[['NorthAmerica_Sales','Europe_Sales','Global_Sales']].mean()
sales_by_Genre

# helps us understand how are different genres contributing to sales, how are they performing in different regions
```
